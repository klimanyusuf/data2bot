```python
schema1 = {
    "type": "object",
    "properties": {
        "name": {"type": "string"},
        "rollnumber": {"type": "number"},
        "marks": {"type": "number"}}}
    
```

```python
def validateJson(f):
    try:
        validate(instance=f, schema=schema1)
    except jsonschema.exceptions.ValidationError as err:
        return False
    return True
```

```python
isValid = validateJson(f)
if isValid:
    print(f)
    print("Given JSON data is True")
else:
    print(f)
    print("Given JSON data is False")
```
